```python
# 23_matplatlib_imshow_all.ipynb

import cv2
import matplotlib.pyplot as plt

#  .
image = cv2.imread( "img/matplot.jpg", cv2.IMREAD_COLOR)

#   ( ) .
rows, cols = image.shape[:2]

#  BGR RGB .
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#   .
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#   figure(, , )  (: 3x4 )
fig = plt.figure(num=3, figsize=(3*3, 4))

# (BGR)  .
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('original(bgr)')
plt.axis('off')  #   .

#  1: RGB  
plt.subplot(1, 3, 2)
plt.imshow(rgb_img)
plt.axis([0, cols, rows, 0])  #   .
plt.title('rgb color')
plt.axis('off')  #   .

#  2:   
plt.subplot(1, 3, 3)
plt.imshow(gray_img, cmap='gray')
plt.title('gray_img2')

plt.tight_layout() #       .
#    .
plt.show()
```
