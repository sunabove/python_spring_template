```python
# 23_matplatlib_imshow.ipynb

import cv2
import matplotlib.pyplot as plt

#  .
image = cv2.imread( "img/matplot.jpg", cv2.IMREAD_COLOR)

#   ( ) .
rows, cols = image.shape[:2]

#  BGR RGB .
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#   .
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#   figure(, , )  (: 3x4 )
fig = plt.figure(num=1, figsize=(3, 4))

# (BGR)  .
plt.imshow(image)
plt.title('figure1- original(bgr)')
plt.axis('off')  #   .
plt.tight_layout() #       .

#   figure  (: 6x4 )
fig = plt.figure(num=2, figsize=(6, 4))
plt.suptitle('figure2- pyplot image display')

#  1: RGB  
plt.subplot(1, 2, 1)
plt.imshow(rgb_img)
plt.axis([0, cols, rows, 0])  #   .
plt.title('rgb color')

#  2:   
plt.subplot(1, 2, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('gray_img2')

#    .
plt.show()
```
