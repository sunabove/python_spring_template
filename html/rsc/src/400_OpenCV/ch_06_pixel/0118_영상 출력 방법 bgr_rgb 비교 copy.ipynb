```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#   (OpenCV  BGR  )
img_bgr = cv2.imread('example.jpg')

# NumPy  BGR RGB 
img_rgb = img_bgr[:, :, ::-1]  #     (BGR -> RGB)
#img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

#    RGB   
plt.figure(figsize=(10, 5))  #   

# BGR   
plt.subplot(1, 2, 1)  # 1 2     
plt.imshow(img_bgr)  # BGR  RGB  
plt.title('BGR Channel Image')
plt.axis('off')

# RGB   
plt.subplot(1, 2, 2)  # 1 2     
plt.imshow(img_rgb)  #  RGB    
plt.title('RGB Channel Image')
plt.axis('off')

#   
plt.tight_layout()
plt.show()
```
