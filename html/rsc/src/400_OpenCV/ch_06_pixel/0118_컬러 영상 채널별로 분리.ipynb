```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#   (OpenCV  BGR  )
img_bgr = cv2.imread('example.jpg')

# NumPy  BGR RGB 
img_rgb = img_bgr[:,:,::-1]  #     (BGR -> RGB)

# RGB  
red_channel   = img_rgb[:, :, 0]    # R 
green_channel = img_rgb[:, :, 1]    # G 
blue_channel  = img_rgb[:, :, 2]    # B 

#          R, G, B  
# R  : R , G B 0 
img_red = np.zeros_like( img_rgb )
img_red[:, :, 0] = red_channel

# G  : G , R B 0 
img_green = np.zeros_like( img_rgb )
img_green[:, :, 1] = green_channel

# B  : B , R G 0 
img_blue = np.zeros_like( img_rgb )
img_blue[:, :, 2] = blue_channel

#      
plt.figure(figsize=(12, 8))

#   (RGB)
plt.subplot(2, 2, 1)
plt.imshow( img_rgb )
plt.title('Original RGB Image')
plt.axis('off')

# Red  
plt.subplot(2, 2, 2)
plt.imshow( img_red )
plt.title('Red Channel')
plt.axis('off')

# Green  
plt.subplot(2, 2, 3)
plt.imshow( img_green )
plt.title('Green Channel')
plt.axis('off')

# Blue  
plt.subplot(2, 2, 4)
plt.imshow( img_blue )
plt.title('Blue Channel')
plt.axis('off')

#    
plt.tight_layout()
plt.show()
```
