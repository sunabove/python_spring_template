```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from collections import OrderedDict  # OrderedDict 

#  () 
image = cv2.imread("example.jpg", cv2.IMREAD_GRAYSCALE)

# OpenCV  
dst1 = cv2.add(image, 100)  #   (saturation )
dst2 = cv2.subtract(image, 100)  #  

# numpy array 
dst3 = image + 100  #   (overflow )
dst4 = image - 100  #   (underflow )

# numpy array 
#   (  ) 
dst5 = np.where( image + 100.0 > 255, 255, image + 100 )
#   (  ) 
dst6 = np.where( image - 100.0 < 0, 0, image - 100 )

#      
images = OrderedDict([
    ('Brightened (OpenCV)', dst1),
    ('Darkened (OpenCV)', dst2),
    ('Brightened (Numpy overflow)', dst3),
    ('Darkened (Numpy underflow)', dst4),
    ('Brightened (Numpy np.where)', dst5),
    ('Darkened (Numpy np.where)', dst6)
])

#    (6  )
chart_size = (rows, cols) = (3, 2)
fig, axes = plt.subplots(rows, cols, figsize=(4*cols, 3*rows))

# axes flatten 1  
axes = axes.flatten()

#       
for i, (title, img) in enumerate(images.items()):
    axes[i].imshow(img, cmap='gray')
    axes[i].set_title(title)
    axes[i].axis('off')  #    axis('off') 
pass

#  
plt.tight_layout()
plt.show()
```
